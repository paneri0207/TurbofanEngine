```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score 
```

```python
data = pd.read_csv("train_FD001.txt" , sep=' ', header=None)
```

```python
data
```

```python
# Drop columns by index positions
data = data.iloc[:, [i for i in range(len(data.columns)) if i not in [26, 27]]]
data
```

```python
data.columns=['Unit_ID', 'Time_Cycle','Opset1','Opset2','SensorMeasure1','SensorMeasure2','SensorMeasure3','SensorMeasure4','SensorMeasure5','SensorMeasure6','SensorMeasure7','SensorMeasure8','SensorMeasure9','SensorMeasure10','SensorMeasure11','SensorMeasure12','SensorMeasure13','SensorMeasure14','SensorMeasure15','SensorMeasure16','SensorMeasure17','SensorMeasure18','SensorMeasure19','SensorMeasure20','SensorMeasure21','SensorMeasure22']
```

```python
data
```

```python

data[data['Unit_ID']==2]
```

```python
#End of cycle = EOL
```

```python
EOL=[]
for sl_no in data['Unit_ID']:EOL.append(((data[data['Unit_ID']==sl_no]["Time_Cycle"]).values)[-1])
```

```python
EOL
```

```python
#Append the EOL column
```

```python
data["EOL"] = EOL
```

```python
data
```

```python
#life ratio 
#Calculate LR
```

```python
data["LR"]=data["Time_Cycle"].div(data["EOL"])
```

```python
data
```

```python

data=data.drop(columns=['Unit_ID', 'EOL'])
```

```python
data
```

```python
labels=[]
for i in range(0,len(data)):
    if np.array(data['LR'])[i]<=0.6:
        labels.append(0)
    elif np.array(data['LR'])[i]<=0.8:
        labels.append(1)
    else :
        labels.append(2)
data['labels']=labels
data=data.drop(columns=['LR'])
data
```

```python
data.to_csv('train_FD001.csv',index=None)
data.to_csv('train_FD002.csv',index=None)
data.to_csv('train_FD003.csv',index=None)
data.to_csv('train_FD004.csv',index=None)
```

```python
df1=pd.read_csv('train_FD001.csv')
df2=pd.read_csv('train_FD002.csv')
df3=pd.read_csv('train_FD003.csv')
df4=pd.read_csv('train_FD004.csv')
```

```python
df=pd.concat([df1,df2,df3,df4])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Assuming df is your dataframe with features and target
X = df.iloc[:, 0:-1]  # Features
y = df.iloc[:, -1]    # Target (last column)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=0)

# Initialize the RandomForestClassifier
RF = RandomForestClassifier()

# Fit the model to the training data
RF.fit(X_train, y_train)  # Ensure this line is typed correctly
```

```python
y_pred_RF = RF.predict(X_test)

cm = pd.DataFrame(confusion_matrix(y_test, y_pred_RF, normalize = 'true')*100)
colormap = sns.color_palette("Reds")
sns.heatmap(cm, annot = True, cmap = colormap)

accuracy_score(y_test, y_pred_RF)
```

```python
RF.get_params()
```

```python
y_pred_RF = RF.predict(X_train)

cm = pd.DataFrame(confusion_matrix(y_train, y_pred_RF, normalize='true')*100)
colormap = sns.color_palette("Blues")
sns.heatmap(cm, annot=True, cmap=colormap)

accuracy_score(y_train, y_pred_RF)
```

```python
rf = RandomForestClassifier()
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, cv = 3, n_jobs=-1)
rf_random.fit(X_train, y_train)
```
